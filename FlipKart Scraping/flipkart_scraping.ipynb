```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
from time import sleep
from random import randint
```

```python
import time
time.sleep(10)
import numpy as np
title=[]

price=[]
review=[]
rating=[]
img=[]
discout=[]
disc=[]
pages = np.arange(1, 1001, 40)
for page in pages:
    url="https://www.flipkart.com/all/pr?sid=all&page=2"
    r=requests.get(url)
    soup=BeautifulSoup(r.text,'html.parser')
    table=soup.find_all('div',{"class":"_13oc-S"})
    
    for i in table[0:41]:
        title.append(i.find_all("a")[1].text.strip())
        
        rating.append(i.find("div",{"class":"_3LWZlK"}).text.strip())
        price.append(i.find("div",{"class":"_30jeq3"}).text.strip())
        review.append(i.find("span",{"class":"_2_R_DZ"}).text.strip())
        discout.append(i.find("div",{"class":"_3Ay6Sb"}).text.strip())
        disc.append(i.find("div",{"class":"_3Djpdu"}).text.strip())
        img.append(i.find("img",{"class":"_396cs4 _3exPp9"}).text.strip())
```

```python
import pandas as pd
data=pd.DataFrame({'Title':title,"Img":img,"Disc":disc,"Rating":rating,"Review":review,"Price":price,"Discount":discout})
data.head() 
```
